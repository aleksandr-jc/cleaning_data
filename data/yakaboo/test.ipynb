```python
import pandas as pd
```

```python
df = pd.read_excel('xlsx/books_yakaboo.xlsx')
df
```

```python
isbn = df[df['ISBN'].notna()]
isbn_right = isbn[isbn['ISBN'].astype(str).str.startswith('978')]
isbn_right
```

```python
isbn_966 = isbn[isbn['ISBN'].astype(str).str.startswith('966')]
isbn_966
```

```python
isbn_200 = isbn[isbn['ISBN'].astype(str).str.startswith('200')]
isbn_200
```

```python
# видаляю ISBN які починаються на 200
df = df.drop(index=isbn_200.index)
```

```python
df.to_csv('csv/books_yakaboo.csv', index=False)
```
