```python
import pandas as pd
```

```python
input_file = 'xlsx/knigoland_fiction.xlsx'

df = pd.read_excel(input_file)
df
```

```python
# вибираємо порожні isbn 
isbn_null = df[df['ISBN'].isnull()]
print(f"Кількість книдок без ISBN: {isbn_null.shape[0]}")
```

```python
isbn_null.head()
```

```python
df = df.drop(index=isbn_null.index)
```

```python
df.shape
```

```python
df.to_excel(input_file, index=False)
```
