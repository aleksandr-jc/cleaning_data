```python
input_file = 'xlsx/knigoland_schoochildren.xlsx'

df = pd.read_excel(input_file)
df.shape
```

```python
df_filter = df.groupby('Жанр книги').count()
df_filter.sort_values(by='Назва книги',ascending=False)
```

```python
df_filter = df[df['ISBN'].isnull()]


x = int(df.shape[0]) - int(df_filter.shape[0])
print(f'Книги в яких є ISBN: {x}')
print(f'Всього книг: {df.shape[0]}')
print(f'Книги без ISBN: {df_filter.shape[0]}')
```

```python
df_filter = df[df['ISBN'].astype(str).str.startswith('978')]
df_filter.shape[0]
```
